```python
word = input("Enter a word: ").lower()

repeat = False

for i in range(len(word)):
    for j in range(i + 1, len(word)):
        if word[i] == word[j]:
            repeat = True

if repeat:
    print("The word has repeating letters")
else:
    print("The word has no repeating letters")
```
